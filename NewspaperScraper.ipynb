```python
from bs4 import BeautifulSoup
import requests
```

# Dataset con 67 registros y tres columnas

```python
# Creando un header
# headers = {'User-agent':'Mozilla/5.0'}
headers = {'User-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

request = requests.get('https://www.milenio.com/', headers=headers)
html = request.content

# Creando soup
soup = BeautifulSoup(html, 'html.parser')
# print(soup.prettify())  # Para buscar los elementos necesarios
```

```python
titles = []
links = []
categories = []

for h in soup.findAll("a", class_="board-module__a"):
    if h.get('target') != '_blank':
        title = h.contents[0]
        link = h['href']
        category = link.split('/')[1]

        titles.append(title)
        links.append(link)
        categories.append(category)

# Código para crear un dataset con las columnas 'title', 'link' y 'category'
```

Este código crea un diccionario llamado data donde las claves son los nombres de las columnas y los valores son las listas correspondientes. Luego, se usa la función pd.DataFrame para crear un dataframe a partir del diccionario. Finalmente, se imprime el dataframe resultante.

```python
import pandas as pd
import numpy as np

data = {'titulo': titles, 'link': links, 'categoria': categories}
df = pd.DataFrame(data)
print(df.shape)
df.sample(10)
```
