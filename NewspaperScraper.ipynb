```python
from bs4 import BeautifulSoup
import requests
```

```python
# Creando un header
# headers = {'User-agent':'Mozilla/5.0'}
headers = {'User-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

request = requests.get('https://www.milenio.com/', headers=headers)
html = request.content

# Creando soup
soup = BeautifulSoup(html, 'html.parser')
# print(soup.prettify())  # Para buscar los elementos necesarios
```

```python
def milenio_news_scrapper(keyword):
  news_list = []

# Buscando todos los headers en la página web de BBC
  for h in soup.findAll("a", class_="board-module__a"):
    news_title = h.contents[0].lower()

    if news_title not in news_list:
      if 'milenio' not in news_title:
        news_list.append(news_title)
  
  no_of_news = 0
  keyword_list = []  # Busca en la lista palabras clave
  for i, title in enumerate(news_list):
    text = ''
    if keyword.lower() in title:
      text = '<-------------- KEYWORD'
      no_of_news += 1
      keyword_list.append(title)
    
    print(i + 1, ':', title, text)

  # Imprimiendo la lista de artículos en los que se incluye la palabra clave
  print(f'\n--------- Menciones totales de "{keyword}" = {no_of_news}--------')
  for i, title in enumerate(keyword_list):
    print(i + 1, ":", title)
```

```python
titles = []
links = []
categories = []

for h in soup.findAll("a", class_="board-module__a"):
    if h.get('target') != '_blank':
        title = h.contents[0]
        link = h['href']
        category = link.split('/')[1]

        titles.append(title)
        links.append(link)
        categories.append(category)

# Código para crear un dataset con las columnas 'title', 'link' y 'category'
```

Este código crea un diccionario llamado data donde las claves son los nombres de las columnas y los valores son las listas correspondientes. Luego, se usa la función pd.DataFrame para crear un dataframe a partir del diccionario. Finalmente, se imprime el dataframe resultante.

```python
import pandas as pd
import numpy as np

data = {'titulo': titles, 'link': links, 'categoria': categories}
df = pd.DataFrame(data)
print(df.shape)
df.tail()
```
